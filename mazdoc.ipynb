```
import cv2
import numpy as np
from matplotlib import pyplot as plt 
from google.colab.patches import cv2_imshow

img = cv2.imread('ocr.jpg', 0)
```

```
binary_img = cv2.Laplacian(img, cv2.CV_64F)
cv2_imshow(binary_img)
```

```
#remove Vertical lines 

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
cv2_imshow(sobely)
```

```
print(sobely.shape)
```

```
#show horizontal lines only

horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (20,1))
show_horizontal = cv2.morphologyEx(sobely, cv2.MORPH_OPEN, horizontal_kernel, iterations = 2)
cv2_imshow(show_horizontal)
```

```
print(show_horizontal.shape)
```

# Vertical Section

```
#remove horizontal lines

laplacian = cv2.Laplacian(img, cv2.CV_64F)
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
cv2_imshow(sobelx)
```

```
#show vertical lines only

vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,5))
show_vertical = cv2.morphologyEx(sobelx, cv2.MORPH_OPEN, vertical_kernel, iterations = 4)
cv2_imshow(show_vertical)
```

```
#blend function

blend = cv2.add(show_vertical , show_horizontal)
cv2_imshow(blend)
```
